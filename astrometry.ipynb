```python
def trimmer(values_ra, values_dec):
    centroid_ra = np.mean(values_ra)
    std_ra = np.std(values_ra)

    centroid_dec= np.mean(values_dec)
    std_dec= np.std(values_dec)

    trimmed_ra = []
    trimmed_dec = []

    for i in range(0,len(values_ra)):
        if abs(values_ra[i]-centroid_ra) <= 3*std_ra and abs(values_dec[i]-centroid_dec) <= 3*std_dec:
            trimmed_ra.append(values_ra[i])
            trimmed_dec.append(values_dec[i])

    return trimmed_ra,trimmed_dec
```

```python
# iterate through all obsids, this will be the one we use as a reference point
obsids = ["23443 lo","23441 lo","23444 hi","21218 hi"]
for primary_obsid in obsids:
    # then iterate through all other obsids, to calculate differences between each pair
    average_obsid_differences = []

    for other_obsid in obsids:
        # then iterate through all point sources in each

        # keep track of differences in RA, DEC, and corresponding values squared
        ra_differences = []
        dec_differences = []

        ra_differences_squared = []
        dec_differences_squared = []
        
        # keep track of error weights as well, which will be used later to calculate average
        ra_errors = []
        dec_errors = []

        for point_source in range(1,13):
            
            # get centroid data for each point source --> Comes out in [RA,DEC,RA_ste,DEC_ste] format
            first_point_source = source_locations[primary_obsid][point_source]
            second_point_source = source_locations[other_obsid][point_source]

            # make sure point source is present in both observations
            if first_point_source[0] != 0 and second_point_source[0] != 0:

                # calculate differences in centroids
                average_dec = (second_point_source[1]+first_point_source[1])/2

                difference_ra = second_point_source[0]-first_point_source[0] 
                difference_dec = second_point_source[1]-first_point_source[1] 

                # sum of square error errors as associated variances for RA and DEC
                var_ra = (first_point_source[2]**2 + second_point_source[2]**2)
                var_dec = (first_point_source[3]**2 + second_point_source[3]**2)

                # account for declination correction
                difference_ra = difference_ra*np.cos(average_dec*(np.pi/180))
                var_ra = var_ra*np.cos(average_dec*(np.pi/180))

                # weight the difference in centroid by the inverse of the variance associated with the difference
                weighted_ra_difference = difference_ra*(1/var_ra)
                weighted_dec_difference = difference_dec*(1/var_dec)

                # also calculate the weighted squared difference in centroid
                weighted_ra_difference_squared = weighted_ra_difference*difference_ra
                weighted_dec_difference_squared = weighted_dec_difference*difference_dec



                # keep track of these values as we append all point sources
                ra_differences.append(weighted_ra_difference)
                dec_differences.append(weighted_dec_difference)
                ra_differences_squared.append(weighted_ra_difference_squared)
                dec_differences_squared.append(weighted_dec_difference_squared)

                # keep track of error weights used
                ra_errors.append(1/var_ra)
                dec_errors.append(1/var_dec)


        # calculate weighted average --> sum of weighted differences / sum of of weights used
        weighted_average_ra_difference = sum(ra_differences)/sum(ra_errors)
        weighted_average_dec_difference = sum(dec_differences)/sum(dec_errors)

        # calculate weighted average squared --> sum of weighted differences squared / sum of of weights used
        weighted_average_ra_difference_squared = sum(ra_differences_squared)/sum(ra_errors)
        weighted_average_dec_difference_squared = sum(dec_differences_squared)/sum(dec_errors)

        # calculated error using Var(weighted difference average) = E(weighted difference average ^2) - E(weighted difference average)^2
        weighted_average_ra_difference_error = (weighted_average_ra_difference_squared-weighted_average_ra_difference**2)**.5
        weighted_average_dec_difference_error = (weighted_average_dec_difference_squared-weighted_average_dec_difference**2)**.5

        # package up the RA, DEC differences and corresponding errors to append to table
        package = (weighted_average_ra_difference,weighted_average_dec_difference,weighted_average_ra_difference_error,weighted_average_dec_difference_error)

        # round for readability, and multiply by 3600 to convert to arcseconds
        package = [np.round(i*3600,8) for i in package]
        average_obsid_differences.append(package)

        print(package)

    all_point_source_differences.loc[len(all_point_source_differences.index)] = [primary_obsid,average_obsid_differences[0],average_obsid_differences[1],average_obsid_differences[2],average_obsid_differences[3]]


    
```
